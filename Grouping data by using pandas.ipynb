# Grouping Data by using Pandas library

**Grouping data by using dataframe.groupby() function**

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/one-piece-arcs/OnePieceArcs.csv')

df.head()
```

```python
ds = df.groupby('Arc')
ds.head()
```

**grouping by using function get_group()**

```python
dx=ds.get_group('Water 7 Arc')
dx.head()
```

**Use groupby() function to form groups based on more than one category (i.e. Use more than one column to perform the splitting).**

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/one-piece-arcs/OnePieceArcs.csv')

gk = df.groupby(['Arc','Manga%'])
gk.head()
```
