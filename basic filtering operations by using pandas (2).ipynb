**Selecting those rows of Pandas Dataframe whose column value is present in the list using isin() method of the dataframe.**

```python
import pandas as pd

ds = pd.read_csv('/kaggle/input/one-piece-arcs/OnePieceArcs.csv')

options = ['Marineford Arc','Return to Sabaody Arc']

rsult_df = ds[ds['Arc'].isin(options)]

rsult_df.head()


```

**Selecting all the rows from the given dataframe in which ‘Stream’ is present in the options list using loc[ ].**

```python
import pandas as pd

ds = pd.read_csv('/kaggle/input/one-piece-arcs/OnePieceArcs.csv')

options = ['Marineford Arc','Return to Sabaody Arc']

rsult_df = ds.loc[ds['Arc'].isin(options)]

rsult_df.head()

```
