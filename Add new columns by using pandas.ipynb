# Add new columns in existing datasets


**Create a Datasets**

```python
import pandas as pd

data = {'Name':['Raj','Aman','Soham'],
       'Age':[22,33,12],
       'Marks':[9,8,7]}
df = pd.DataFrame(data)
df.head()
```

**Create a new column in Existing Dataframe Using a List**

```python
city = ['Pune','Mumbai','Nagpur']
df['city'] = city

df.head()
```

```python
data = {'Aman': pd.Series([1,2,3,4]),
        'Soham': pd.Series([23,43,23,45])}
df = pd.DataFrame(data)
df.head()
```

```python
Rishi = [12,31,12,43]
df['Rishi'] = Rishi
df.head()
```

**Creating a new Column in Exisiting DataFrame Using df.insert()**

```python
df.insert(1,'Nikita',[12,43,23,21],True)
df.head()
```

**Creating a new Column in Existing DataFrame Using df.assign()**

```python
df.assign(Rahul = [12,32,45,21])
```

**Creating a New Column in Existing DataFrame Using a Dictonary**

```python
data = {'Name':['Ronit','Sai','Jaideep'],
       'Age':[22,21,22],
       'Marks':[56,76,78]}
df = pd.DataFrame(data)
df.head()
```

```python
percentage = {98:'Shivam',
             90:'Raju',
             97:'Rohan'}
df['percentage'] = percentage
df.head()
```
