# Sorting data from Datasets


```python
import pandas as pd

ds = pd.read_csv('/kaggle/input/one-piece-arcs/OnePieceArcs.csv')
df = ds.sort_values(by = 'Arc')
df.head()
```

**Sorting values by dscending order**


```python
import pandas as pd 

ds = pd.read_csv('/kaggle/input/one-piece-arcs/OnePieceArcs.csv')
df = ds.sort_values(by = 'Arc',ascending = False)
df.head()
```

**Sorting Data based on sampling**


```python
import pandas as pd

ds= pd.read_csv('/kaggle/input/one-piece-arcs/OnePieceArcs.csv')
df = ds.sort_values(by='Arc',na_position='first')
df.head()
```

**Sorting values by using multiple columns**
